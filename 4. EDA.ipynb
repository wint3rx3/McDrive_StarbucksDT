```python
import pandas as pd

mcdonalds_df = pd.read_csv('data/맥도날드_상권정보.csv', encoding='utf-8')
starbucks_df = pd.read_csv('data/스타벅스_상권정보.csv', encoding='utf-8')
```

```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster

# 지도 설정 (중심: 대한민국)
map_center = [36.5, 127.5]  # 대한민국의 대략적인 중심 좌표

# 1. 맥도날드 지도
mcdonalds_map = folium.Map(location=map_center, zoom_start=7)
for _, row in mcdonalds_df.iterrows():
    folium.Marker(
        location=[row['lat'], row['lng']],
        popup=f"McDonald's: {row['name']}\nAddress: {row['address']}",
        icon=folium.Icon(color='red', icon='cutlery')
    ).add_to(mcdonalds_map)

# 2. 스타벅스 지도
starbucks_map = folium.Map(location=map_center, zoom_start=7)
for _, row in starbucks_df.iterrows():
    folium.Marker(
        location=[row['lat'], row['lng']],
        popup=f"Starbucks: {row['name']}\nAddress: {row['address']}",
        icon=folium.Icon(color='green', icon='star')
    ).add_to(starbucks_map)

# 3. 맥도날드 + 스타벅스 지도
combined_map = folium.Map(location=map_center, zoom_start=7)
# 맥도날드 추가
for _, row in mcdonalds_df.iterrows():
    folium.Marker(
        location=[row['lat'], row['lng']],
        popup=f"McDonald's: {row['name']}\nAddress: {row['address']}",
        icon=folium.Icon(color='red', icon='cutlery')
    ).add_to(combined_map)
# 스타벅스 추가
for _, row in starbucks_df.iterrows():
    folium.Marker(
        location=[row['lat'], row['lng']],
        popup=f"Starbucks: {row['name']}\nAddress: {row['address']}",
        icon=folium.Icon(color='green', icon='star')
    ).add_to(combined_map)

# 각각의 지도를 저장
combined_map
```
